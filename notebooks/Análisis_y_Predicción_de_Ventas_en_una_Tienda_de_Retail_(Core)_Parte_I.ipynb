EXPLORACIÓN DE DATOS

```python
# VENTAS TOTALES POR CATEGORÍA

electronics = df[df['Product Category'] == 'Electronics']
venta_electronics = electronics['Total Amount'].sum()
print(f'Ventas totales Electronics: {venta_electronics}')

beauty = df[df['Product Category'] == 'Beauty']
venta_beauty = beauty['Total Amount'].sum()
print(f'Ventas totales Beauty: {venta_beauty}')

clothing = df[df['Product Category'] == 'Clothing']
venta_clothing = clothing['Total Amount'].sum()
print(f'Ventas totales Clothing: {venta_clothing}')
```

```python
## ALTERNATIVA USANDO GROUPBY
df.groupby('Product Category')['Total Amount'].sum()
```

```python
# PROMEDIO DE VENTAS DIARIAS POR CATEGORÍA USANDO GROUPBY

df.groupby('Product Category')['Total Amount'].mean()
```

```python
# PRODUCTOS CON MAYOR Y MENOR VENTAS

mayor_ventas = (df.groupby('Product Category')['Total Amount'].sum()).idxmax()    # Método .idxmax() devuelve el nombre de la columna (o sea el índice) donde se encuentra
                                                                                  # el valor máximo
print(f'El producto con mayor ventas corresponde a la categoría de: {mayor_ventas}')

menor_ventas = (df.groupby('Product Category')['Total Amount'].sum()).idxmin()
print(f'El producto con menor ventas corresponde a la categoría de: {menor_ventas}')
```

MANIPULACIÓN DE DATOS

```python
# Mostrar solo las ventas de una categoría de producto específica
# Se elige trabajar con la categoria 'Electronics' por ser solamente la de mayor peso en la tienda al ser la que más vende

electronics = df[df['Product Category'] == 'Electronics'].copy()    # Se crea una copia para no modificar el df original


# OPERACIONES A REALIZAR:
# Se cobrará un iva, se hará descuentos especiales aplicados al precio final y se crearán nuevas columnas para operar.

# Se crea un cargo por iva del 19%
electronics['IVA (19%)'] = electronics['Total Amount'] * 0.19

# Se actualiza el precio con iva incluido
electronics['Total (IVA inlcuido)'] = electronics ['Total Amount'] + electronics['IVA (19%)']

# Suponiendo descuento de la tienda de un 20% (se aplica sobre el producto, no el valor final)
electronics['Descuento promo'] = electronics['Total (IVA inlcuido)'] - electronics['Total Amount'] * 0.02


```
