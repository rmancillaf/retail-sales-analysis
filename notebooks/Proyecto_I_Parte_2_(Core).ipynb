5. FILTRADO DE DATOS

```python
# Filtrar por ventas mayores a 50

###COMENTARIOS: No existe una columna llamada 'Ventas', sin embargo en análisis de datos tiene más sentido que éste término se refiera al valor monetario (dinero)
###             y no a cuántas piezas se vendieron. Para este caso, aquella columna con 'Total Amount' es la que mejor representa el concepto de 'Ventas'.
###             Se procederá a renombrar la columna de 'Total Amount' como 'Ventas', sin antes hacer una copia del df para evitar sobreescribir el df original.

df_copia = df.copy()
df1_copia = df_copia.rename(columns = {'Total Amount': 'Ventas'})
df1_copia[df1_copia['Ventas'] > 50]
```

```python
# Filtrar por 'Precio' menor a 0.5

###COMENTARIOS: El nombre 'Precio' NO EXISTE en el dataset. Sin embargo hay una categoría que representa muy bien esta variable y es la columna 'Price per Unit'.
###             Por lo tanto, como en el caso anterior, se renombrará la columna 'Price per Unit' por 'Precio' entendiendo que ambos conceptos son similares. Pero,
###             surge otro problema, y es que el precio mínimo, como se aprecia en las estadísticas descriptivas, es de 30 unidades monetarias y al usar 0.5
###             indicado en las instrucciones el resultado será 'vacío'. Por ende se reemplazará el valor de 0.5 por 50, valor totalmente al azar
###             pero dentro del rango de precios.

df2_copia = df1_copia.rename(columns = {'Price per Unit': 'Precio'})
df2_copia[df2_copia['Precio'] < 50]
```

```python
# Utilizando query() para filtrar el df y que muestre las filas donde 'Producto' sea 'Manzanas' y 'Ventas' sean mayores a 30

###COMENTARIOS: Columna 'Producto' se mencionó que no existía, y como lo más cercano a esta nombre es 'Product Category', se realiza el método .unique()
###             a ésta columna donde no hay rastros de la palabra 'Manzanas'. Como se verificó que la cantidad de Non-Null count = N° de registros (no hay NaN),
##              con dtypes se mostró aquellas columnas con valores tipo 'object', que es donde probablemente se encuentre la palabra 'Manzanas', dado que el resto de
###             columnas son números enteros, entonces se procede a usar el método .str.contains() para comprobar finalmente que 'Manzanas' no se encuentra
###             en el dataset. Como la idea es entender las metodologías, 'Manzanas' será reemplazada por la categoría de producto
###             'Clothing' por el hecho solamente de ser la categoría con la mayor cantidad de valores únicos que podría representar mayor demanda,
###             entendiéndose que no hay data que represente el producto 'Manzanas' y por tanto el resultado sería 'vacío'.



print(df['Product Category'].unique())
print()
df2_copia.query("`Product Category` == 'Clothing' & Ventas > 30")   # Product Category tiene espacios y Python pensará que Product y Category son dos cosas distintas
                                                                    # dando error de sintaxis. NOMBRES DE COLUMNAS CON ESPACIOS VAN ENTRE COMILLAS INVERSAS (backticks)
```

#### ***COMENTARIO ADICIONAL***: ABAJO SE ENCUENTRA EL MÉTODO .str.contains() PARA DEMOSTRAR QUE EN LAS CATEGORÍAS DONDE HAY PRESENCIA DE DATOS TIPO OBJECT, NINGUNA COINCIDE CON LA PALABRA 'Manzanas'

```python
df_manzanas = df[
    df['Customer ID'].str.contains('Manzanas') |
    df['Gender'].str.contains('Manzanas') |
    df['Product Category'].str.contains('Manzanas')
]

print(f"Cantidad de palabras 'Manzanas': {len(df_manzanas)}")
```

6. SLICING DE DATOS

```python
# Mostrar columnas 'Producto' y 'Ventas' del df

###COMENTARIOS: Columna 'Producto' será reemplazada por 'Product Category' para observar solamente cómo se ejecuta el slicing.

df2_copia[['Product Category', 'Ventas']]
```

```python
# Utilizar método loc[] para mostrar las filas 5-10 inclusive, y columnas 'Producto' y 'Tienda'

###COMENTARIOS: Mismo procedimiento que se ha trabajado en este CORE. Recordar que para fines ilustrativos 'Producto' será 'Product Category'
###             y 'Tienda' será 'Gender' como se hizo en ejercicios previos, pero ambos conceptos no existen en el df.

df2_copia.loc[5:10, ['Product Category', 'Gender']]
```
