```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/retail_sales_dataset.csv')
df
```

TRANSFORMACIÓN DE DATOS

```python
df_copia = df.copy()
```

```python
# Se crearán nuevas columnas como total a pagar con el iva

df_copia['Tax Amount'] = df_copia['Total Amount'] * 0.19
df_copia['Final Price'] = df_copia['Total Amount'] + df_copia['Tax Amount']
df_copia
```

```python
# Normalizando las ventas

# Esto permite comparar por ejemplo el rendimiento de las ventas en una escala de 0 a 1
max_value = df_copia['Final Price'].max()
min_value = df_copia['Final Price'].min()
df_copia['Sales Normalized'] = df_copia['Final Price'].apply(lambda x: (x - min_value) / (max_value - min_value))
df_copia
```

```python
# Categorizando las ventas

# Se traen las estadísticas descriptivas para utilizar los valores y quizás definir una función

df_copia.describe()
```

```python
# Categorizando las ventas podríamos usar los cuantiles (.quantile()) para que haya una distribución justificada y no arbitraria

precio_bajo = df_copia['Final Price'].quantile(0.25)
precio_alto = df_copia['Final Price'].quantile(0.75)

def clasificar_venta(venta):
  if venta <= precio_bajo:
    return 'Low'
  elif venta >= precio_alto:
    return 'High'
  else:
    return 'Medium'

# Aplicando apply para que se aplique la función a cada columna del precio final
df_copia['Sales Category'] = df_copia['Final Price'].apply(clasificar_venta)
df_copia
```

AGRUPACIÓN Y AGREGACIÓN

```python
# Agrupación de categoría por producto y género. Esto puede ayudar en análisis de datos para revelar preferencias en consumo según segmento demográfico, en este caso,
# según el género de la persona

df_copia.groupby(['Product Category', 'Gender'])['Final Price'].agg(
    ['count', 'sum', 'mean', 'min', 'max', 'std', 'var']).rename(columns = {'sum': 'Total Sales', 'mean': 'Average Sale', 'min': 'Lowest Price', 'max': 'Max. Price'})
```

Existe una alta preferencia por la categoría vesturario y electrónica, se consume y gasta bastante en estos sectores y que tiene sentido en un mundo donde la tecnología prima. Por otro lado en la vestimenta, la ropa básica se considera necesidad básica (protección contra el clima, salud) pero la moda o el lujo son más secundarias ya que mejoran la calidad de vida o social pero no son vitales para la supervivencia, sin embargo, de igual forma se gasta bastante en este sector ya sea para mantener un alto status social, autoestima, satisfacción personal, etc. Para los productos de belleza, existe una gran preferencia del cuidado e higiene personal por parte de las mujeres vs los hombres

```python
# Análisis personalizado: se utilizará una función para calcular cuánto se aleja cada venta respecto del promedio global
# Esto permite conocer si la transacción fue normal o extraordinaria

prom_global = df_copia['Final Price'].mean()

# Se define la función 'calc_resultado' que devolverá la diferencia del valor de la columna 'Final Price' y el promedio global

def calc_resultado(valor):
  resultado = valor - prom_global
  # Uso de condicional para determinar el rendimiento de la venta
  if resultado > 0:
    return 'Outperforming🔥'
  elif resultado == 0:
    return 'On Target✅'
  else:
    return 'Underperforming❌'

# Se aplica la función a cada columna del precio final y se devuelve a una nueva columna de rendimiento
df_copia['Transaction Rating'] = df_copia['Final Price'].apply(calc_resultado)
df_copia
```
